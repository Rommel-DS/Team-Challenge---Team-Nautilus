```python
import random
import numpy as np

class Tablero:
    def __init__(self, dimensiones=(10, 10), id_user=1, id_machine=0):
        self.dimensiones = dimensiones
        self.id_user = id_user
        self.id_machine = id_machine
        self.tablero_user = np.full(dimensiones, " ")
        self.tablero_machine = np.full(dimensiones, " ")
        self.barcos_user = {}
        self.barcos_machine = {}

    def set_tablero(self, barcos_dict, tablero):
        # Coloca los barcos dados en el tablero especificado
        for barco, posiciones in barcos_dict.items():
            for pos in posiciones:
                fila, columna = pos
                if 0 <= fila < self.dimensiones[0] and 0 <= columna < self.dimensiones[1]:
                    tablero[fila, columna] = "O"
                else:
                    print(f"Advertencia: La posición {pos} de {barco} está fuera del tablero y se ignorará.")

    def crear_barcos_aleatorios(self, tablero, esloras=[1, 2, 3], cantidades=[4, 3, 2], num_intentos=100):
        barcos_dict = {}
        
        for eslora, cantidad in zip(esloras, cantidades):
            for n in range(cantidad):
                barco = []
                contador = 0
                while True:
                    if contador >= num_intentos:
                        print(f"Límite de intentos superado para barco de eslora {eslora}")
                        break
                    contador += 1
                    
                    # Coordenadas de la posición inicial
                    fila_inicial = random.randint(0, self.dimensiones[0] - 1)
                    columna_inicial = random.randint(0, self.dimensiones[1] - 1)
                    
                    # Orientación aleatoria
                    orientacion = random.choice(["N", "S", "O", "E"])
                    posiciones = [(fila_inicial, columna_inicial)]
                    
                    for i in range(1, eslora):
                        if orientacion == "N":
                            fila_inicial -= 1
                        elif orientacion == "S":
                            fila_inicial += 1
                        elif orientacion == "E":
                            columna_inicial += 1
                        elif orientacion == "O":
                            columna_inicial -= 1
                        
                        if 0 <= fila_inicial < self.dimensiones[0] and 0 <= columna_inicial < self.dimensiones[1]:
                            posiciones.append((fila_inicial, columna_inicial))
                        else:
                            break
                    
                    if len(posiciones) == eslora and all(tablero[fila, columna] == " " for fila, columna in posiciones):
                        for fila, columna in posiciones:
                            tablero[fila, columna] = "O"
                        barcos_dict[f"barco{eslora}_{n+1}"] = posiciones
                        break
                    else:
                        print("Intentando colocar otro barco de eslora", eslora)
        
        return barcos_dict

    def disparar(self, fila, columna, tablero_oponente):
        if tablero_oponente[fila, columna] == "O":
            print("¡Impacto!")
            tablero_oponente[fila, columna] = "X"  # Marca el impacto
            return "impacto"
        else:
            print("Agua")
            tablero_oponente[fila, columna] = "A"  # Marca el agua
            return "agua"
    
    def iniciar_juego(self):
        # Inicializa los tableros del usuario y la máquina con barcos aleatorios
        self.barcos_user = self.crear_barcos_aleatorios(self.tablero_user)
        self.barcos_machine = self.crear_barcos_aleatorios(self.tablero_machine)
        print("Tablero del usuario:")
        print(self.tablero_user)
        print("Tablero de la máquina (escondido):")
        print(self.tablero_machine)
```

```python
juego = Tablero()
juego.iniciar_juego()
```
